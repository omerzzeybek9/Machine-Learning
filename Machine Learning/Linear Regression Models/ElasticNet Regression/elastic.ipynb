```python
#Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge, Lasso , ElasticNet
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split,cross
from sklearn import model_selection
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
```

```python
df = pd.read_csv("Hitters.csv")

df = df.dropna()

dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])

y = df["Salary"]

X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype("float64")

X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
enet_model = ElasticNet().fit(X_train,y_train)
```

```python
enet_model.coef_
```

```python
enet_model.intercept_
```

```python
enet_model.predict(X_train)[0:10]
```

```python
enet_model.predict(X_test)[0:10]
```

```python
y_pred = enet_model.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
rmse
```

```python
r2_score(y_test,y_pred)
```

```python
alphas = 10**np.linspace(10,-2,100)*0.5
l1_ratios = np.linspace(0,1,11)
l1_ratios
```

```python
enet_cv = ElasticNetCV(alphas=alphas,l1_ratio=l1_ratios,cv=10).fit(X_train,y_train)
```

```python
enet_cv.alpha_
```

```python
enet_cv.l1_ratio_
```

```python
enet_cv.coef_
```

```python
enet_cv.intercept_
```

```python
enet_tuned = ElasticNet(alpha=enet_cv.alpha_,l1_ratio=enet_cv.l1_ratio_).fit(X_train,y_train)
```

```python
y_pred = enet_tuned.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
rmse
```
